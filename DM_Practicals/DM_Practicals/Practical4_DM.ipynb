```python
df = pd.read_csv('bake.csv')
```

```python
df.columns=["RN","Food","Qt"]
```

```python
df['Food'].replace({1:"milk",2:"sugar",3:"chocolate",4:"apples",5:"curd"},inplace=True)
```

```python
df=df.drop(columns=['Qt'])
```

```python
df=pd.get_dummies(df)
```

```python
df=df.groupby('RN')[['Food_apples','Food_chocolate','Food_curd','Food_milk','Food_sugar']].sum()
```

```python
df['Food_apples'].loc[df['Food_apples']>0]='apples'
df['Food_chocolate'].loc[df['Food_chocolate']>0]='chocolate'
df['Food_curd'].loc[df['Food_curd']>0]='curd'
df['Food_milk'].loc[df['Food_milk']>0]='milk'
df['Food_sugar'].loc[df['Food_sugar']>0]='sugar'
```

```python
df
```

```python
tran2 = df.values.tolist()
```

```python
for i in range(1000):
    if 0 in tran2[i]:
        tran2[i].remove(0)
```

```python
tran2
```

```python
is1,r1 = apriori(tran2, min_support=0.02,min_confidence=0.075)
```

```python
print(is1)
```

```python
for rule in r1:
  print(rule)
```
